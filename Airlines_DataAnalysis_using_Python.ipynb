```python
#ANALYSIS STARTS FROM HERE

#1. How many flights have 100+ seats in it?

pd.read_sql_query("""SELECT aircraft_code, count(*) as total_seats FROM seats
                    GROUP BY aircraft_code
                    HAVING total_seats > 100 """, conn)
```

```python
#2. How many bookings received and revenue earned between the timed/said months?
# plotting this using line-graph

tickets = pd.read_sql_query("""SELECT * FROM tickets
                    inner join bookings
                    on tickets.book_ref = bookings.book_ref""", conn)



tickets['book_date'] = pd.to_datetime(tickets['book_date'])
# Creating a new column DATE that will only have the date and not the timezone in it. 
tickets['only_date'] = tickets['book_date'].dt.date
tickets
```

```python
# Using GroupBy function from the Pandas (Another way apart from SQL groupby)
tickets.groupby('only_date')[['only_date']].count()
```

```python
# Plotting line-Graph

x = tickets.groupby('only_date')[['only_date']].count()
plt.figure(figsize = (20, 6))
plt.plot(x.index, x['only_date'], marker = "^")
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Number  of   Tickets', fontsize = 20)
plt.grid('b')
plt.show()
```

```python
bookings = pd.read_sql_query("""SELECT * FROM bookings""", conn)
bookings['book_date'] = pd.to_datetime(bookings['book_date'])

# Creating a new column DATE that will only have the date and not the timezone in it. 
bookings['only_date'] = bookings['book_date'].dt.date
bookings
```

```python
# Using GroupBy function from the Pandas (Another way apart from SQL groupby)
bookings.groupby('only_date')[['total_amount']].sum()
```

```python
# Plotting line-Graph

x = bookings.groupby('only_date')[['total_amount']].sum()
plt.figure(figsize = (20, 6))
plt.plot(x.index, x['total_amount'], marker = "^")
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Total Amount of  Tickets', fontsize = 20)
plt.grid('b')
plt.show()
```

```python
#3. Calcuate the average charges for each flight with variable fare rates?
# plotting this using line-graph

df = pd.read_sql_query("""SELECT fare_conditions, aircraft_code, avg(amount)
                            FROM ticket_flights
                            inner join flights
                            on ticket_flights.flight_id = flights.flight_id
                            GROUP BY aircraft_code, fare_conditions""", conn)
df
```

```python
# Plotting Bar-Graph

sns.barplot(data = df, x = 'aircraft_code', y = 'avg(amount)', hue = 'fare_conditions')
```

```python
# Analyzing Occupancy Rate Section.

#1. For each flight, calculate total business revenue per year and the average revenue per ticket

pd.read_sql_query("""SELECT aircraft_code, ticket_count, total_revenue, total_revenue/ticket_count as avg_revenue_per_ticket
                    FROM
                        (SELECT aircraft_code, count(*) as ticket_count, sum(amount) as total_revenue 
                        FROM ticket_flights
                        JOIN flights
                        ON ticket_flights.flight_id = flights.flight_id
                        GROUP BY aircraft_code)""", conn)

```

```python
#2. Calculate average occupancy per aircraft

# pd.read_sql_query("""SELECT aircraft_code, flights.flight_id, count(*) as seats_count 
#                         FROM boarding_passes
#                         JOIN flights
#                         ON boarding_passes.flight_id = flights.flight_id
#                         GROUP BY aircraft_code, flights.flight_id""", conn)   # Query 1 as a


# pd.read_sql_query("""SELECT aircraft_code, count(*) as num_seats_count 
#                         FROM seats
#                         GROUP BY aircraft_code""", conn)                      # Query 2 as b


occupancy_rate = pd.read_sql_query("""SELECT a.aircraft_code, avg(a.seats_count) as booked_seats, b.num_seats_count, 
                                      avg(a.seats_count)/b.num_seats_count as occupancy_rate
                                      FROM
                                         (SELECT aircraft_code, flights.flight_id, count(*) as seats_count 
                                          FROM boarding_passes
                                          JOIN flights
                                          ON boarding_passes.flight_id = flights.flight_id
                                          GROUP BY aircraft_code, flights.flight_id) as a
                                              INNER JOIN
                                                     (SELECT aircraft_code, count(*) as num_seats_count 
                                                     FROM seats
                                                     GROUP BY aircraft_code) as b
                                              ON a.aircraft_code = b.aircraft_code
                                              GROUP BY a.aircraft_code
                                              """, conn)
occupancy_rate
```

```python
#3. Calculate by how much the total annual turnover could increase by giving all aircrafts a 10% higher occupancy rate

occupancy_rate['Inc_occupany_rate'] = occupancy_rate['occupancy_rate'] + occupancy_rate['occupancy_rate']*0.10
occupancy_rate
```

```python
total_revenue = pd.read_sql_query("""SELECT aircraft_code, sum(amount) as total_revenue
                                          FROM ticket_flights
                                          JOIN flights
                                          ON ticket_flights.flight_id = flights.flight_id
                                          GROUP BY aircraft_code""", conn)
total_revenue
```

```python
occupancy_rate['Inc_total_annual_turnover'] = (total_revenue['total_revenue']/occupancy_rate['occupancy_rate'])*occupancy_rate['Inc_occupany_rate']
occupancy_rate
```
